# Nearest Neighbor Analysis To Analyze NSCLC Cell Embeddings

```python
# loading packages
import pandas as pd
from sklearn.neighbors import NearestNeighbors
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

Loading the four embeddings that were created using different methods. Each set of embeddings represents the 20K NSCLC cells in a lower dimensional space.

```python
# load PCA embeddings
path_pca_embeddings = "/home/cog/nlandzaat/rep_learning_sc_foundation/analysis/nlandzaat/4. output/PCA_embeddings_NSCLC_256.csv"
embeddings_PCA = pd.read_csv(path_pca_embeddings)

# load UMAP embeddings
path_umap_embeddings = "/home/cog/nlandzaat/rep_learning_sc_foundation/analysis/nlandzaat/4. output/UMAP_embeddings_NSCLC.csv"
embeddings_umap = pd.read_csv(path_umap_embeddings)

# load GF embeddings
path_gf_embeddings = "/home/cog/nlandzaat/rep_learning_sc_foundation/analysis/nlandzaat/4. output/GF_embeddings_NSCLC.csv"
embeddings_GF = pd.read_csv(path_gf_embeddings)


# load scGPT embeddings
path_scGPT_embeddings = "/home/cog/nlandzaat/rep_learning_sc_foundation/analysis/nlandzaat/4. output/scGPT_embeddings_NSCLC.csv"
embeddings_scGPT = pd.read_csv(path_scGPT_embeddings)
```

# NN

The provided code initializes nearest neighbor instances with 20 neighbors and the Euclidean distance metric. Subsequently, the embeddings are fitted to these instances, enabling the identification of the 20 nearest neighbors for each point using .kneighbors(). 

The resulting distances array stores the distances from each point to its closest neighbors, while the indices array contains the indices of these nearest neighbors.

```python
k_neighbors = 20
```

```python
# create NN objects and distances PCA

# initialize NearestNeighbors objects
nbrs_PCA = NearestNeighbors(n_neighbors=k_neighbors, metric="euclidean")

# fit the embeddings to the NearestNeighbors objects
nbrs_PCA.fit(embeddings_PCA)

# find nearest neighbors for each point in the PCA embeddings
distances_PCA, indices_PCA = nbrs_PCA.kneighbors(embeddings_PCA)
```

```python
# create NN objects and distances UMAP

# initialize NearestNeighbors objects
nbrs_umap = NearestNeighbors(n_neighbors=k_neighbors, metric="euclidean")

# fit the embeddings to the NearestNeighbors objects
nbrs_umap.fit(embeddings_umap)

# find nearest neighbors for each point in the UMAP embeddings 
distances_umap, indices_umap = nbrs_umap.kneighbors(embeddings_umap)
```

```python
# create NN objects and distances GF

# initialize NearestNeighbors objects
nbrs_GF = NearestNeighbors(n_neighbors=k_neighbors, metric="euclidean")

# fit the embeddings to the NearestNeighbors objects
nbrs_GF.fit(embeddings_GF)

# find nearest neighbors for each point in the Geneformer embeddings
distances_GF, indices_GF = nbrs_GF.kneighbors(embeddings_GF)
```

```python
# create NN objects and distances scGPT

# initialize NearestNeighbors objects
nbrs_scGPT = NearestNeighbors(n_neighbors=k_neighbors, metric="euclidean")  

# fit the embeddings to the NearestNeighbors objects
nbrs_scGPT.fit(embeddings_scGPT)

# find nearest neighbors for each point in the scGPT embeddings 
distances_scGPT, indices_scGPT = nbrs_scGPT.kneighbors(embeddings_scGPT)
```

# Analysis

The subsequent code cells conduct a pairwise assessment of the four embeddings. For each pair, the percentage of matching nearest neighbors is computed, alongside the mean percentage and its standard deviation. Additionally, the number of cells with 0 and 20 matching neighbors is determined. Lastly, the percentage of total cells corresponding to each count of matching neighbors (ranging from 0 to 20) is calculated.

### PCA & UMAP

```python
# PCA & UMAP

# calculate the percentage of matching neighbors between the PCA and UMAP embeddings for each pair of cell embeddings
percentages_1 = []
for idx_pca, idx_umap in zip(indices_PCA, indices_umap):
    common_neighbors = np.intersect1d(idx_pca, idx_umap)
    percentage = len(common_neighbors) / len(idx_pca) * 100
    percentages_1.append(percentage)

# calculate the mean and standard deviation of the percentages
mean_percentage_pca_umap = np.mean(percentages_1)
std_dev_percentage_pca_umap = np.std(percentages_1)

# count the number of cells with 0 matching neighbors and 20 matching neighbors (100%)
no_matching_pca_umap = np.sum(np.array(percentages_1) == 0)
all_matching_pca_umap = np.sum(np.array(percentages_1) == 100)
```

```python
# find common neighbors between PCA and UMAP embeddings for each pair of corresponding indices
matching_nn_pca_umap = []
for idx_pca, idx_umap in zip(indices_PCA, indices_umap):
    common_neighbors = np.intersect1d(idx_pca, idx_umap)
    count_common_nn = len(common_neighbors)
    matching_nn_pca_umap.append(count_common_nn)

# count occurrences of each count of matching neighbors (0 to 20)
matching_nn_counts = {i: matching_nn_pca_umap.count(i) for i in range(21)}

# calculate total number of cells
total_cells = len(matching_nn_pca_umap)

# initialize an empty list to store the percentages
percentages_list_pca_umap = []

# calculate and output the percentage for each count of matching neighbors
for count, occurrences in matching_nn_counts.items():
    percentage = (occurrences / total_cells) * 100
    percentages_list_pca_umap.append(percentage)
```
